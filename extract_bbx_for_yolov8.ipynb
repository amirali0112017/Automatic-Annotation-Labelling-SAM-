```python
import json

def extract_bounding_boxes(json_file):
    with open(json_file) as f:
        data = json.load(f)

    images = data['images']
    annotations = data['annotations']

    for image in images:
        image_id = image['id']
        file_name = image['file_name']
        width = image['width']
        height = image['height']

        image_annotations = [ann for ann in annotations if ann['image_id'] == image_id]

        # Create a text file for the bounding boxes
        txt_file = file_name.split('.')[0] + '.txt'

        with open(txt_file, 'w') as f:
            for annotation in image_annotations:
                bbox = annotation['bbox']
                x = bbox[0]
                y = bbox[1]
                bbox_width = bbox[2]
                bbox_height = bbox[3]

                # Normalize the bounding box coordinates
                x /= width
                y /= height
                bbox_width /= width
                bbox_height /= height

                # Write the bounding box coordinates to the text file
                f.write(f'0 {x} {y} {bbox_width} {bbox_height}\n')

        print(f"Bounding boxes for {file_name} extracted and saved in {txt_file}.")

# Provide the path to your JSON file
json_file_path = r'E:\Amir\EWTracker\dataset_cropped\2\New folder\2.json'

# Call the function to extract bounding boxes
extract_bounding_boxes(json_file_path)
```

```python
# Go through the txt files and find the number of frames with four earthworms detected,three earthworms detected, two earthworms detected, one earthworm detected, and zero earthworms detected.


import os
import glob

# Get the list of all txt files in the current directory
txt_files = glob.glob('*.txt')

# Initialize the counts
four_earthworms = 0
three_earthworms = 0
two_earthworms = 0
one_earthworm = 0
zero_earthworms = 0

for txt_file in txt_files:
    with open(txt_file, 'r') as f:
        lines = f.readlines()
        num_lines = len(lines)

        if num_lines == 4:
            four_earthworms += 1
        elif num_lines == 3:
            three_earthworms += 1
        elif num_lines == 2:
            two_earthworms += 1
        elif num_lines == 1:
            one_earthworm += 1
        elif num_lines == 0:
            zero_earthworms += 1

print(f"Number of frames with four earthworms detected: {four_earthworms}")
print(f"Number of frames with three earthworms detected: {three_earthworms}")
print(f"Number of frames with two earthworms detected: {two_earthworms}")
print(f"Number of frames with one earthworm detected: {one_earthworm}")
print(f"Number of frames with zero earthworms detected: {zero_earthworms}")



```
